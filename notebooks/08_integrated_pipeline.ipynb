```python
import sys
from pathlib import Path

project_root = Path().resolve().parent 
sys.path.append(str(project_root))

from src.classification.classifier import predict
from src.rag.pipeline import answer
```

```python
def run_pipeline(query: str) -> dict:
    """
    Full integrated pipeline:
    query → classify → retrieve → generate → answer
    """
    # Step 1: classify the query
    category = predict(query)
    
    # Step 2: get RAG answer
    result = answer(query)
    
    # Step 3: add category to the result
    result["category"] = category
    result["question"] = query
    
    return result
```

```python
test_queries = [
    # Symptoms (1)
    "What are the early symptoms of type 2 diabetes?",
    # Diagnosis (2)
    "How is pneumonia diagnosed in elderly patients?",
    # Treatment (3)
    "What are the current treatment options for hypertension?",
    # Medication (4)
    "What are the side effects of metformin?",
    # Prevention (5)
    "How can cardiovascular disease be prevented through lifestyle changes?",
    # General (6)
    "What is the role of the immune system in fighting infections?",
    # Extra queries
    "Are there advantages of induction therapy in renal transplantation?",
    "What does cryotherapy do before drainage?",
    "What are the symptoms of diaphragmatic herniation?",
    "How does suprascapular nerve entrapment present clinically?",
]
```

```python
import json

results = []
for q in test_queries:
    r = run_pipeline(q)
    results.append(r)
    print(f"Q: {q[:60]}...")
    print(f"   Category: {r['category']}")
    print(f"   Answer preview: {r['answer'][:120]}...")
    print(f"   Disclaimer present: {r['disclaimer_present']}")
    print()
```

```python
with open("reports/integrated_pipeline_test_results.json", "w") as f:
    json.dump(results, f, indent=2)

print(f"Saved {len(results)} results.")
```
